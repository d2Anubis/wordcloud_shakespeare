```python
import pandas as pd 
import numpy as np
from wordcloud import WordCloud
from nltk.corpus import stopwords 
import matplotlib.pyplot as plt 
from PIL import Image
```

```python
train=pd.read_csv("./shakespeare_data.csv")
```

```python
words = '' 
stopwords = set(stopwords.words('english'))
```

```python
for val in train.PlayerLine: 
    val = str(val) 
    tokens = val.split() 
    for i in range(len(tokens)): 
        tokens[i] = tokens[i].lower() 
    words += " ".join(tokens)+" "
  
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = stopwords, 
                min_font_size = 10).generate(words) 
```

```python
plt.figure(figsize = (8,8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show() 
```

```python
img= np.array(Image.open("./william-shakespeare-black-silhouette.jpg"))
img
```

```python
wc = WordCloud(background_color="white", max_words=1000, mask=img,
               stopwords=stopwords, contour_width=3, contour_color='black')
wc.generate(words)
wc.to_file("./final.png")
plt.figure(figsize=[20,10])
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```
